```python
def calcular_returns(date, df, es_stock):
    f = dateutil.parser.parse(date)            
    Qini = 0
    maxIntentos = 1            
    returns = 0
    while Qini==0 and maxIntentos < 30:
        if (es_stock):
            fini = (str(f.month - 2)).zfill(2) + '-' + (str(maxIntentos)).zfill(2) + '-' + str(f.year)
        else:
            fini = str(f.year) + '-' + (str(f.month - 2)).zfill(2) + '-' + (str(maxIntentos)).zfill(2)
        try:
            if (es_stock):
                dd = df[df['DATE'] == fini]
                Qini = dd['PRICE'].values[0]
            else:
                Qini = df.loc[fini]['Open']        
        except:
            pass
        maxIntentos = maxIntentos + 1
    Qfin = 0
    maxIntentos = 31 if f.month == 3 or f.month==12 else 30
    while Qfin == 0 and maxIntentos > 1:
        if (es_stock):
            ffin = (str(f.month)).zfill(2) + '-' + str(maxIntentos).zfill(2) + '-' + str(f.year)
        else:
            ffin = str(f.year) + '-' + (str(f.month)).zfill(2) + '-' + str(maxIntentos).zfill(2)
        try:
            if (es_stock):
                dd = df[df['DATE'] == ffin]        
                Qfin = dd['PRICE'].values[0]
            else:
                Qfin = df.loc[ffin]['Close']        
        except:
            pass
        maxIntentos = maxIntentos - 1      
    try :
        if Qini == 0:
            returns = 0 #Ha habido una excepcion porque no encontraba el precio para ese día
        else:
            returns = (Qfin - Qini) / Qini
    except:
        returns = 0
    if returns == -np.inf:
        returns = 0
    return returns, Qini, Qfin
```

```python
%%time
financialDF['RETURNS'] = 0
tickers_no_calculados = []
#¡¡NO EJECUTAR CON TODAS LAS COMPAÑIAS!!
for c in lista_companias:    
#for c in ['AAPL']:
    try:        
        df = financialDF.loc[(c),]        
        list_returns = []
        list_returns_index = []
        list_prices_open = []
        list_prices_close = []
        price_company = prices.loc[(c),]
        for date in df.index:            
            returns, qini, qfin = calcular_returns(date, price_company, True)
            returns_index, _, _, = calcular_returns(date, index, False)
            #No he conseguido que me actualice financialDF, pongo esto:
            #financialDF.loc[(date),'RETURNS'] = returns
            #y no me actualiza la columna returns, sin embargo con esto si actualiza df
            #df.loc[(date),'RETURNS'] = returns
            #df.loc[(date),'INDEX_RETURNS'] = returns_index
            list_returns.append(returns)
            list_returns_index.append(returns_index)
            list_prices_open.append(qini)
            list_prices_close.append(qfin)
        df['PRICE_INI'] = np.array(list_prices_open)
        df['PRICE_FIN'] = np.array(list_prices_close)
        df['RETURNS'] = np.array(list_returns)        
        df['INDEX_RETURNS'] = np.array(list_returns_index)
        col_to_shift1 = ['RETURNS']
        col_to_shift2 = ['INDEX_RETURNS']
        df['RETURNS_NEXT'] = df[col_to_shift1].shift(-1)
        df['INDEX_RETURNS_NEXT'] = df[col_to_shift2].shift(-1)
        df['RETURNS_CAT'] = np.where(df['RETURNS']>0, 'POSITIVE', 'NEGATIVE')
        df['RETURNS_NEXT_CAT'] = np.where(df['RETURNS_NEXT']>0, 'POSITIVE', 'NEGATIVE')
        df['RETURNS_vs_INDEX'] = np.where(df['RETURNS']>df['INDEX_RETURNS'], 'WINS', 'LOSES')
        df['RETURNS_vs_INDEX_NEXT'] = np.where(df['RETURNS_NEXT']>df['INDEX_RETURNS_NEXT'], 'WINS', 'LOSES')
        df['Ticker'] = c
        filename = 'Datos_csv/'+c+'.csv'
        df.to_csv(filename, sep=';')
        print('Ticker',c,'calculado')
    except:
        tickers_no_calculados.append(c)
        pass
```
